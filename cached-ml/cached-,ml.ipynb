```python
import pandas
import numpy
import sklearn.linear_model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
sklearn.linear_model.LinearRegression
```

```python
n = 1000
n2 = 50
x = numpy.random.rand(n)
x2 = numpy.random.rand(n2)
y = x * (2 + numpy.random.rand(n)) + numpy.random.rand(n) * 0.05
y2 = numpy.random.rand(n2) * 3
```

```python
df = pandas.concat(
    [pandas.DataFrame({"x": x, "y": y, "category": [0]*len(x)}),
     pandas.DataFrame({"x": x2, "y": y2, "category": [1]*len(x2)})
    ], axis=0
).reset_index(drop=True)
```

```python
train, test = train_test_split(df, test_size=0.5)
```

```python
def plot(df):
    c0 = df[df["category"] == 0]
    c1 = df[df["category"] == 1]
    plt.scatter(c0["x"], c0["y"], c="b")
    plt.scatter(c1["x"], c1["y"], c="r")
```

```python
plot(train)
```

```python
plot(test)
```
