```python
import pandas
import numpy
```

```python
df = pandas.read_csv("dataset/feature_normalized.csv")
train_df = df[15800:23200].reset_index(drop=True)
test_df = df[25000:35000].reset_index(drop=True)
pred = pandas.read_csv("test_df.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
pred["Profit"] = test_df["Profit"]
pred["Date"] = test_df["Date"]
```

```python
pred["cap_ghold"] = pred["Profit"].cumsum() + 1
```

```python
n = 24 * 14
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(pred["cap"][:n], c="r", label="reinforcement_learning")
ax1.plot(pred["cap_ghold"][:n], c="b", label="ghold")
ax2 = ax1.twinx()
ax2.plot(pred["btc/jpy"][:n], c="g", label="btc/jpy")
ax1.legend()
ax2.legend()
plt.savefig("eval.png")
```
